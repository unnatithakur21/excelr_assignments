```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("C:/Users/krishna/Downloads/books.csv",sep=',',encoding= 'unicode_escape')
data
```

```python
#rename the Solar column
book = data.rename({'User.ID': 'userid','Book.Title':'title','Book.Rating':'rating'}, axis=1)
book1=book.drop(['Unnamed: 0'],axis=1)
book1.head()
```

```python
book1[book1.duplicated()].shape
```

```python
book2=book1.drop_duplicates()
book2
```

```python
book2.isnull().sum() #null values
```

```python
#number of unique users in the dataset
len(book2.userid.unique())
```

```python
#number of unique users in the dataset
len(book2.title.unique())
```

```python
#number of unique users in the dataset
len(book2.rating.unique())
```

```python
book3 = book2.pivot_table(index='userid', columns='title',
                                 values='rating').reset_index(drop=True)
```

```python
book3
```

```python
book3.index = book2.userid.unique()
```

```python
book3.head()
```

```python
#Impute those NaNs with 0 values
book3.fillna(0, inplace=True)
book3.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book3.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book2.userid.unique()
user_sim_df.columns = book2.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
