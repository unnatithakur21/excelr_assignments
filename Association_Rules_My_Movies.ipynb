```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("C:/Users/krishna/Downloads/my_movies.csv")
movies
```

```python
movies1=movies.iloc[:,5:]
movies1
```

# Apriori algorithm

# Keeping minimum support = 10%

```python
frequent_itemsets = apriori(movies1, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
#Using Lift as Metric, keeping min value as 1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'], rules['confidence'],c=rules['lift'],cmap= 'summer',alpha=0.75)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
cbar=plt.colorbar()
cbar.set_label('lift')
```

```python
#Using Confidence as Metric, keeping min value as 75%
rules2=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.75)
rules2
```

```python
plt.figure(figsize=(10,5))
import matplotlib.pyplot as plt
plt.scatter(rules2['support'], rules2['confidence'],c=rules2['lift'],cmap= 'summer',alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
cbar=plt.colorbar()
cbar.set_label('lift')
```

```python
#Using Supprort as Metric, keeping min value as 50%%
rules3 = association_rules(frequent_itemsets, metric="support", min_threshold=0.5)
rules3.sort_values('lift',ascending = False)
```

```python
plt.figure(figsize=(10,5))
import matplotlib.pyplot as plt
plt.scatter(rules3['support'], rules3['confidence'],c=rules3['lift'],cmap= 'summer',alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
cbar=plt.colorbar()
cbar.set_label('lift')
```
